```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Substitua pelo caminho correto do seu arquivo CSV
file_path = '/home/davioliveira/projects/FIBRA-Larces/visualizer/execution_times.csv'  
data = pd.read_csv(file_path)

# Criar DataFrame com os dados
tempo_inserir_network_traffic = data['tempo_inserir_network_traffic']
tempo_deteccao_conexao = data['tempo_detecção_conexao']

df = pd.DataFrame({
    "tempo_inserir_network_traffic": tempo_inserir_network_traffic,
    "tempo_deteccao_conexao": tempo_deteccao_conexao
})
```

```python
df.replace(0, np.nan, inplace=True)

# Exibir o DataFrame com os valores alterados
print(df)
```

```python
# Remover as linhas que contêm NaN
df.dropna(inplace=True, ignore_index=True)

# Exibir o DataFrame com as linhas sem NaN
print(df)
```

```python
# Calculando média de tempo pra inserir na network-traffic
mean_time = df['tempo_inserir_network_traffic'].mean()
print(mean_time)
```

```python
# Configurações do Seaborn para estilo
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))

# Gráfico de linha para os valores
sns.lineplot(data=df, x=range(len(df)), y="tempo_inserir_network_traffic", marker="o", label="Inserção network-traffic", color="blue")
sns.lineplot(data=df, x=range(len(df)), y="tempo_deteccao_conexao", marker="s", label="Detecção conexão", color="orange")

# Configurações do gráfico
plt.title("Tempo para operações de conexão", fontsize=16)
plt.xlabel("Índice", fontsize=12)
plt.ylabel("Tempo (ms)", fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
