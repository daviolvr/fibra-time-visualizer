```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Substitua pelo caminho correto do seu arquivo CSV
file_path = '/home/davioliveira/projects/FIBRA-Larces/visualizer/execution_times.csv'  
data = pd.read_csv(file_path)

# Criar DataFrame com os dados
src_tempo_aplicar_regras_whitelist = data['src_tempo_aplicar_regras_whitelist']
dst_tempo_aplicar_regras_whitelist = data['dst_tempo_aplicar_regras_whitelist']

df= pd.DataFrame({
    "tempo_aplicar_regras_whitelist": pd.concat([data['src_tempo_aplicar_regras_whitelist'], data['dst_tempo_aplicar_regras_whitelist']], ignore_index=True)
})

print(df)
```

```python
df.replace(0, np.nan, inplace=True)

# Remover as linhas que contêm NaN
df.dropna(inplace=True, ignore_index=True)

# Exibir o DataFrame com as linhas sem NaN
print(df)
```

```python
# Calculando média de tempo pra aplicar regras para whitelist
mean_time = df['tempo_aplicar_regras_whitelist'].mean()
print(mean_time)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))

# Gráfico de linha para os valores
sns.lineplot(data=df, x=range(len(df)), y="tempo_aplicar_regras_whitelist", marker="o", label="Tempo de aplicar regras para whitelist", color="blue")

# Configurações do gráfico
plt.title("Tempo para aplicar regras para whitelist", fontsize=16)
plt.xlabel("Índice", fontsize=12)
plt.ylabel("Tempo (ms)", fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
