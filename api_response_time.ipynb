```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Substitua pelo caminho correto do seu arquivo CSV
file_path = '/home/davioliveira/projects/FIBRA-Larces/visualizer/execution_times.csv'  
data = pd.read_csv(file_path)

# Criar DataFrame com os dados
src_tempo_resposta_api = data['src_tempo_resposta_api']
dst_tempo_resposta_api = data['dst_tempo_resposta_api']

df = pd.DataFrame({
    "tempo_resposta_api": pd.concat([data['src_tempo_resposta_api'], data['dst_tempo_resposta_api']], ignore_index=True)
})

print(df)
```

```python
df.replace(0, np.nan, inplace=True)

# Remover as linhas que contêm NaN
df.dropna(inplace=True, ignore_index=True)

# Exibir o DataFrame com as linhas sem NaN
print(df)
```

```python
# Calculando média de tempo pra API retornar uma resposta
mean_time = df['tempo_resposta_api'].mean()
print(mean_time)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 5))

# Gráfico de linha para os valores
sns.lineplot(data=df, x=range(len(df)), y="tempo_resposta_api", marker="o", label="Tempo de resposta da API", color="blue")

# Configurações do gráfico
plt.title("Tempo para API retornar uma resposta", fontsize=16)
plt.xlabel("Índice", fontsize=12)
plt.ylabel("Tempo (ms)", fontsize=12)
plt.legend(fontsize=12)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
